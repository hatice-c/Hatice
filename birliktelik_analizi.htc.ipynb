```python
data = list(df['products'].apply(lambda x:x.split(',')))
data
```

```python
!pip install mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
te = TransactionEncoder()
te_data = te.fit(data).transform(data)
df = pd.DataFrame(te_data,columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import apriori
```

```python
freq_items=apriori(df,min_support=0.15,use_colnames=True,verbose=1)
freq_items.head()
```

```python
association_rules(freq_items,metric="confidence",min_threshold=0.3)
```

**1. Maggi- Bread Karsilastirmasi:** Beraber alinma yuzdesi %15. Maggi alindiginda Bread'inde alinma yuzdesi %60. Yani maggi alinan her 100 alisverisin 60 inda Bread de alinmistir.

**2. Maggi-Tea Karsilastirmasi:** Beraber alinma yuzdesi %20. Maggi alindiginda Tea'nin alinma yuzdesi %80. Her Maggi alinan 100 alisverisin 80ninde Tea de alinmistir

**3. Milk- Bread Karsilastirmasi:** Beraber alinma yuzdesi %20. Milk alindiginda Bread'in alinma yuzdesi %80. Her Milk alinan 100 alisverisin 80ninde Bread de alinmistir.

***Durum Analizi :***
    Bu market icin satislarin tumune baktigimizda ekmek alinma orani %65dir. Sepete en cok giren ekmektir. Ekmek satislarina bagli olarak ekmegin birlikte sepete en cok beraber girdigi urunler %80 lik oranla Maggi ve Milk tir. Bread Maggi ve Milk in standlarini magazanin farkli kesimlerine dagitarak musterilerin daha fazla gezinip alisveris yapmalari saglanabilir. 
    Maggi alanlarin %80inin Tea aldigini da biliyoruz. Maggi, Milk, Tea ve Bread i birbirinden farkli yerlere konumlandirmak mantikli bir secim olacaktir
